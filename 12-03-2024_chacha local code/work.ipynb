```python
import matplotlib.pyplot as plt

# Algorithm names
algorithms = ['NLP', 'Logistic Regression', 'Random Forest', 'SVM']

# Metrics
precision = [0.85, 0.78, 0.81, 0.79]
recall = [0.82, 0.75, 0.79, 0.77]
f1_score = [0.83, 0.76, 0.80, 0.78]
accuracy = [0.96, 0.80, 0.84, 0.82]

# Colors for each metric
colors = ['blue', 'green', 'orange', 'red']

# Plotting
plt.figure(figsize=(10, 6))

# Precision
plt.plot(algorithms, precision, marker='o', label='Precision', color=colors[0])

# Recall
plt.plot(algorithms, recall, marker='o', label='Recall', color=colors[1])

# F1-score
plt.plot(algorithms, f1_score, marker='o', label='F1-score', color=colors[2])

# Accuracy
plt.plot(algorithms, accuracy, marker='o', label='Accuracy', color=colors[3])

# Adding labels and title
plt.xlabel('Algorithms')
plt.ylabel('Score')
plt.title('Comparison of Algorithms')
plt.legend()

# Displaying the plot
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
